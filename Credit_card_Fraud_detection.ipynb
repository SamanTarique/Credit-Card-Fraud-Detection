# **Step 6   Model Selection**

## **1) Introduction**

Model Selection means *choosing the best machine learning algorithm* for your problem.
For fraud detection, we need a model that can:

* Handle imbalanced data
* Detect rare fraud transactions
* Give high accuracy, precision, and recall

## **2) Why Model Selection is Important**

Different models behave differently:
| Model               | Pros                               | Cons                             |
| ------------------- | ---------------------------------- | -------------------------------- |
| Logistic Regression | Simple, interpretable              | May not capture complex patterns |
| Random Forest       | Handles non-linear data, robust    | Slower on large datasets         |

| Decision Tree       | Easy to understand                 | Can overfit                      |

We will select **multiple models**, compare their performance, and choose the best one.
## **3) Handling Imbalanced Data**

Fraud cases are rare → imbalanced dataset.
Options:

1. *Resampling Techniques*

   * **SMOTE**: Synthetic Minority Oversampling Technique
   * Creates synthetic fraud examples

2. *Class Weight in Model*

   * Some models can give more weight to fraud class

We will use *SMOTE* for resampling.

### **Code**

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

print("Resampled feature shape:", X_res.shape)
print("Resampled target shape:", y_res.shape)
```

## **4) Selecting Candidate Models**

We choose 3 models to try:

* *Logistic Regression*
* *Random Forest Classifier*
* *XGBoost Classifier*

These are popular for fraud detection.

###  **Code**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

lr_model = LogisticRegression(
    max_iter=2000,
    random_state=42
)

rf_model = RandomForestClassifier(random_state=42)

xgb_model = XGBClassifier(
    eval_metric='logloss',
    random_state=42
)
```


## **5) Splitting Data for Training and Testing**

Before training, split data into *train* and *test* sets.

###  **Code**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res,
    test_size=0.2,
    random_state=42,
    stratify=y_res
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```


## **6) Summary of Model Selection**

* Imbalanced data handled with **SMOTE**
* Candidate models: Logistic Regression, Random Forest, XGBoost
* Data split into **training** and **testing** sets

#  **Step 7  - Model Training**

## **1) Definition of Model Training**

**Model Training** is the process where a machine learning algorithm learns patternsfrom the dataset.

* Input: Features (X)
* Output: Target labels (y)

During training, the model **adjusts its internal parameters** to predict correctly.
In fraud detection, training allows the model to **understand the difference between normal and fraud transactions**.



## **2) Importance of Model Training**

* Ensures the model can **predict new unseen transactions**
* Converts data into a **learned model** that recognizes fraud patterns
* The better the training, the more accurate the predictions
* Training with **balanced data** is important to avoid bias toward normal transactions



## **3) Training Process**

Steps involved in model training:

1. Use **training dataset** only (X_train, y_train)
2. Fit models (Logistic Regression, Random Forest, XGBoost)
3. Save trained models for evaluation/testing
4. Optionally, tune hyperparameters for better performance



## **4) Training Logistic Regression**

###  **Code**







```python
from sklearn.preprocessing import StandardScaler

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Logistic Regression on SCALED data
lr_model.fit(X_train_scaled, y_train)

# Predict
y_train_pred_lr = lr_model.predict(X_train_scaled)

print("Logistic Regression training complete!")
```


## **5) Training Random Forest**

###  **Code**





```python
# Random Forest
rf_model.fit(X_train, y_train)

# Predict on training data
y_train_pred_rf = rf_model.predict(X_train)

print("Random Forest training complete!")
```



## **6) Training XGBoost**

###  **Code**





```python
# XGBoost
xgb_model.fit(X_train, y_train)

# Predict on training data
y_train_pred_xgb = xgb_model.predict(X_train)

print("XGBoost training complete!")
```



## **7) Saving Trained Models (Optional)**

After training, we can save models to reuse without retraining.

###  **Code**





```python
import joblib

joblib.dump(lr_model, 'lr_model.pkl')
joblib.dump(rf_model, 'rf_model.pkl')
joblib.dump(xgb_model, 'xgb_model.pkl')
```



## **8) Summary of Model Training**

#  **Step 8 – Model Evaluation & Tuning**



## **1) Definition of Model Evaluation**

**Model Evaluation** means checking how well a trained model performs on **unseen (test) data**.
It helps us understand:

* How accurate the model is
* How many fraud cases are correctly detected
* How many normal transactions are wrongly marked as fraud

In fraud detection, **accuracy alone is not enough**, because fraud cases are rare.



## **2) Why Evaluation is Important**

Evaluation helps to:

* Compare different models
* Select the best performing model
* Avoid overfitting
* Improve model reliability

Important metrics for fraud detection:

* **Precision** → How many detected frauds are actually fraud
* **Recall** → How many real frauds are caught
* **F1-Score** → Balance between precision and recall
* **Confusion Matrix** → Shows correct and wrong predictions



## **3) Model Evaluation Metrics**

###  **Code (Import Metrics)**




```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```


## **4) Evaluate Logistic Regression**

###  **Code**




```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_test_pred_lr = lr_model.predict(X_test_scaled)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_test_pred_lr))
print(classification_report(y_test, y_test_pred_lr))
print(confusion_matrix(y_test, y_test_pred_lr))
```


## **5) Evaluate Random Forest**

###  **Code**







```python
y_test_pred_rf = rf_model.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_test_pred_rf))
print(classification_report(y_test, y_test_pred_rf))
print(confusion_matrix(y_test, y_test_pred_rf))
```



## **6 Evaluate XGBoost**

###  **Code**





```python
y_test_pred_xgb = xgb_model.predict(X_test)

print("XGBoost Accuracy:", accuracy_score(y_test, y_test_pred_xgb))
print(classification_report(y_test, y_test_pred_xgb))
print(confusion_matrix(y_test, y_test_pred_xgb))
```
